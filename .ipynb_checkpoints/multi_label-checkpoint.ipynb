```python
# opening the labels data file
df=pd.read_csv("./miml_dataset/miml_labels_1.csv")
df.head()
```

```python
# Recording the label names
LABELS=["desert", "mountains", "sea", "sunset", "trees"]
```

```python
# Building Image File List Dataset
data_dir = pathlib.Path("miml_dataset")
filenames = list(data_dir.glob('images/*.jpg'))
print("image count: ", len(filenames))
print("first image: ", str(filenames[0]) )
PIL.Image.open(str(filenames[0]))
```

```python
# Creating a list of image file paths
fnames=[]
for fname in filenames:
  fnames.append(str(fname))
fnames[:5]
```

```python
# Testing the input pipeline with a small portion of the data to ensure the performance and accuracy of the tf.data 
#pipeline design.

# Let's select a very small set of images and observe the effects of map, prefect, cache, and batch methods.

# Let's supply the image file list to tf.data.Dataset.from_tensor_slices() for creating a dataset of file paths.

ds_size= len(fnames)
print("Number of images in folders: ", ds_size)

#number_of_selected_samples=10
number_of_selected_samples=ds_size
filelist_ds = tf.data.Dataset.from_tensor_slices(fnames[:number_of_selected_samples])

ds_size= filelist_ds.cardinality().numpy()
print("Number of selected samples for dataset: ", ds_size)
```

```python
# 4. Prepareing Auxiliary Functions
# Let's start by checking if the filelist_ds is created as expected by getting some samples
print("Number of files in the dataset: ", filelist_ds.cardinality().numpy())
print("3 samples:")
for a in filelist_ds.take(3):
  fname= a.numpy().decode("utf-8")
  print(fname)
  display(PIL.Image.open(fname)) 
```

```python
# 4.2 Get the label for the given image file name
# Remember we read the label info into a pandas DF
df.head()
```

```python
# Coding a simple function get_label(file_path) to match the file name with the corresponding one-hot encoded 
# labels as shown below:
def get_label(file_path):
  print("get_label acivated...")
  parts = tf.strings.split(file_path, '/')
  file_name= parts[-1]
  labels= df[df["Filenames"]==file_name][LABELS].to_numpy().squeeze()
  return tf.convert_to_tensor(labels)
```

```python
# Let's check the function:
for a in filelist_ds.take(5):
  print("file_name: ", a.numpy().decode("utf-8"))
  print(get_label(a).numpy())
```

```python
# 4.3 Process the image
# Let's process (resize, scale, etc.) the images so that we can save time in training in the process_img() function
IMG_WIDTH, IMG_HEIGHT = 224, 224
def process_img(img):
  print("process_img acivated...")
  #color images
  img = tf.image.decode_jpeg(img, channels=3) 
  #convert unit8 tensor to floats in the [0,1]range
  img = tf.image.convert_image_dtype(img, tf.float32) 
  #resize 
  return tf.image.resize(img, [IMG_WIDTH, IMG_HEIGHT])
```

```python
# 4.4. Combine the label and image
# We use the combine_images_labels() function:

# to upload the image data from the given file path
# to process the image
# to get the one-hot encoded labels
# to combine image data (X_train) with lablel (y_train) to create the train_ds
def combine_images_labels(file_path: tf.Tensor):
  print("combine_images_labels acivated...")
  img = tf.io.read_file(file_path)
  print("image file read...")
  img = process_img(img)
  label = get_label(file_path)
  return img, label
```

```python
# 5. Decide Split Ratio
# Let's decide the size of train, test and validate datasets and split the file names accordingly to them
train_ratio = 0.80
ds_train=filelist_ds.take(ds_size*train_ratio)
#ds_test_validate=filelist_ds.skip(ds_size*train_ratio)
ds_test_validate=filelist_ds.skip(ds_size*train_ratio)
ds_validate = ds_test_validate.take(ds_test_validate.cardinality().numpy() * 0.5)
ds_test = ds_test_validate.skip(ds_test_validate.cardinality().numpy() * 0.5)


print("train size: ", ds_train.cardinality().numpy())
print("test size: ", ds_test.cardinality().numpy())
print("validate size: ", ds_validate.cardinality().numpy())
```

```python
# 6. Build the train and test datasets
# In train and test datasets, we will have:

# image data
# label data
# So far, we have only the file paths and names stored in these datasets.

ds_train.element_spec
```

```python
# Note that we have only a single tensor of file path (tf.string).
```

```python
# 6.1. map(map_func) method
# Maps map_func across the elements of this dataset.

# We will use map method for calling lambda function to consume each file paths and names 
# and return us (image,label) pairs.
# Let's use tf.py_function, which allows you to write arbitrary Python code

ds_train=ds_train.map(lambda x: tf.py_function(func=combine_images_labels,
          inp=[x], Tout=(tf.float32,tf.int64)),
          num_parallel_calls=tf.data.AUTOTUNE,
          deterministic=False)
ds_train.prefetch(ds_size-ds_size*train_ratio)
```

```python
# 6.2.0 Improve map() performance
#"Performance can often be improved by setting num_parallel_calls so that map will use multiple threads to 
#process elements. If deterministic order isn't required, it can also improve performance to 
#set deterministic=False."

#Since reading files from hard disk takes much time, we apply these suggested methods.
ds_test= ds_test.map(lambda x: tf.py_function(func=combine_images_labels,
          inp=[x], Tout=(tf.float32,tf.int64)),
          num_parallel_calls=tf.data.AUTOTUNE,
          deterministic=False)
ds_test.prefetch(ds_size-ds_size*train_ratio)
```

```python
# Let's check the element spec of the resulting datasets
ds_train.element_spec
```

```python
# 6.2.1 Improve map() performance
#"Performance can often be improved by setting num_parallel_calls so that map will use multiple threads to 
#process elements. If deterministic order isn't required, it can also improve performance to 
#set deterministic=False."

#Since reading files from hard disk takes much time, we apply these suggested methods.
ds_validate= ds_validate.map(lambda x: tf.py_function(func=combine_images_labels,
          inp=[x], Tout=(tf.float32,tf.int64)),
          num_parallel_calls=tf.data.AUTOTUNE,
          deterministic=False)
ds_validate.prefetch(ds_size-ds_size*train_ratio)
```

```python
def covert_onehot_string_labels(label_string,label_onehot):
  labels=[]
  for i, label in  enumerate(label_string):
     if label_onehot[i]:
       labels.append(label)
  if len(labels)==0:
    labels.append("NONE")
  return labels
covert_onehot_string_labels(LABELS,[0,1,1,0,1])
```

```python
def show_samples(dataset):
  fig=plt.figure(figsize=(16, 16))
  columns = 3
  rows = 3
  print(columns*rows,"samples from the dataset")
  i=1
  for a,b in dataset.take(columns*rows): 
    fig.add_subplot(rows, columns, i)
    plt.imshow(np.squeeze(a))
    plt.title("image shape:"+ str(a.shape)+" ("+str(b.numpy()) +") "+ 
              str(covert_onehot_string_labels(LABELS,b.numpy())))
    i=i+1
  plt.show()
show_samples(ds_train.take(2))
```

```python
# 7. Configure Datasets for Training & Testing
# 7.1. Decide Batch Size
BATCH_SIZE=4
```

```python
# 7.2. Configure for performance
# As discussed in Better performance with the tf.data API, there are several ways to increase the performance of input data pipelines in tf.data.

# By increasing the performance of the input pipeline, we can decrease the overall train and test processing time.

# The general strategy to overlap the input pipeline steps (reading file paths, uploading and processing image data, filtering lable info, converting them to image data and label tuples, etc.) with batch computation during train or test phase.

# Here, I summarize the most related methods very shortly:

# Prefetching: Prefetching overlaps the preprocessing and model execution of a training step.
# Caching: The tf.data.Dataset.cache transformation can cache a dataset, either in memory or on local storage. This will save some operations (like file opening and data reading) from being executed during each epoch.
# Batching: Combines consecutive elements of this dataset into batches.
# The exact order of these transformation depends on several factor. For more details please refer to Better performance with the tf.data API

# Below, I will apply these transformations step by step in different orders
```

```python
#.batch(BATCH_SIZE).cache().prefetch(tf.data.experimental.AUTOTUNE)
ds_train_batched=ds_train.batch(BATCH_SIZE).cache().prefetch(tf.data.experimental.AUTOTUNE) 
ds_test_batched=ds_test.batch(BATCH_SIZE).cache().prefetch(tf.data.experimental.AUTOTUNE)
ds_validate_batched=ds_validate.batch(BATCH_SIZE).cache().prefetch(tf.data.experimental.AUTOTUNE)
```

```python
print("Number of batches in train: ", ds_train_batched.cardinality().numpy())
print("Number of batches in test: ", ds_test_batched.cardinality().numpy())
print("Number of batches in validate: ", ds_validate_batched.cardinality().numpy())
```

```python
def plot_hist(hist, metric):
    if metric == 'auc_3':
        plt.plot(hist.history["auc_3"])
        plt.plot(hist.history["val_auc_3"])
    else:
        plt.plot(hist.history["loss"])
        plt.plot(hist.history["val_loss"])
    plt.style.use("ggplot")
    plt.title("model {}".format(metric))
    plt.ylabel("{}".format(metric))
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()
```

```python
#Transfer Learning
#Feature Extraction Step
```

```python
def build_model():
    # Building the original base EfficientNet-0 model
    baseModel = EfficientNetB0(weights="imagenet", 
                  include_top=False, # make sure top layer i.e. head is not included
                  input_tensor=Input(shape=(224, 224, 3)),
                  pooling="avg"
                  )
    
    # freeze the weights
    for layer in baseModel.layers:
        layer.trainable = False
    
    # training a new classifier on top
    x = baseModel.output
    Layer_1 = BatchNormalization()(x)
    Layer_2 = Dropout(0.5)(Layer_1)
    output_layer = Dense(len(LABELS), activation="softmax")(Layer_2)
    
    model = Model(inputs = baseModel.input, outputs = output_layer)
    
    # compile
    opt = Adam(learning_rate=1e-3)
    model.compile(optimizer=opt, 
                loss='binary_crossentropy', 
                metrics=[tf.keras.metrics.AUC()])
    
    return model
```

```python
model_preprocess = build_model() 
```

```python
# implementing early stopping

es = EarlyStopping(monitor='val_loss',  #metric to monitor
                   mode='min',  # whether to minimize or maximize the metric monitored
                   patience=10,  # how many epochs to wait before declaring stopped training
                   verbose=1  # output epoch when training was stopped
                   )
```

```python
# model checkpoint - save the best model with best accuracy on val set
mc = ModelCheckpoint('saved_model/my_model',#'feature_extraction.h5', 
                     monitor='val_loss', 
                     mode='min', 
                     verbose=1, # will tell the epoch+ accuracy everytime model is saved
                     save_best_only=True)
```

```python
# Training the model
hist = model_preprocess.fit(
    x=ds_train_batched,
    epochs=25,
    verbose=2,
    #validation_data=valGen,
    validation_data=ds_validate_batched,
    # shuffle=True, # argument is ignored when `x` is a generator
    # steps_per_epoch=totalTrain // BATCH_SIZE,
    callbacks=[es, mc]
)
```

```python
# plotting losses to see if it would have benefitted from more epochs

plot_hist(hist, metric="loss")
```
